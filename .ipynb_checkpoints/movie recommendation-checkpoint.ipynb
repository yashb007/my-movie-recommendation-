```python
ratings = pd.DataFrame(final.groupby('title').mean()['rating'])
```

```python
ratings["no of ratings"] = pd.DataFrame(final.groupby('title').count()['rating'])
```

```python
final.head()
```

```python
mat = final.pivot_table(index='user_id',columns='title',values='rating')
```

```python
mat.head()
```

```python
ratings.sort_values('no of ratings',ascending=False)
```

```python
starwar_user_rating = mat['Star Wars (1977)']
starwar_user_rating
```

```python
similiar_To_starwars = mat.corrwith(starwar_user_rating)
```

```python
similiar_To_starwars
```

```python
corr_with_starwars = pd.DataFrame(similiar_To_starwars, columns=['Correlation'])
```

```python
corr_with_starwars.dropna(inplace=True)
```

```python
corr_with_starwars.sort_values('Correlation',ascending=False).head(20)
```

```python
corr_with_starwars = corr_with_starwars.join(ratings['no of ratings'])
```

```python
corr_with_starwars
```

```python
corr_with_starwars = corr_with_starwars[corr_with_starwars['no of ratings']>100]
```

```python
corr_with_starwars = corr_with_starwars.sort_values('Correlation',ascending = False)
```

```python
corr_with_starwars
```

# Predict Type

```python
def prediction(movie):
        movie_user_rating = mat[movie]
        similiar_To_movie = mat.corrwith(movie_user_rating)
        
        corr_movie = pd.DataFrame(similiar_To_movie, columns=["Correlation"])
        corr_movie.dropna(inplace=True)
        
        corr_movie = corr_movie.join(ratings['no of ratings'])
        pred = corr_movie[corr_movie['no of ratings']>100].sort_values('Correlation',ascending = False)
        
        return pred
```

```python
p = prediction("Crash (1996)")
```
